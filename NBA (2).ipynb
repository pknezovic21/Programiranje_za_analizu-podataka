**Statistika igrača po draft pickovima**

U idućoj analizi ćemo provjeriti da li stvarno postoji značajna razlika u draft pozicijama sa tzv. lutrijskim pickovima (pickovi 1-15 unutar prve runde). Podijelit ćemo lutrijske pickove na 3 skupa: (1-5), (6-10) i (11-15) te ćemo usporediti njihove prosjeke.

```python

def categorize_draft_pick(overall_pick):
    if overall_pick <= 5:
        return 'Top 5'
    elif 6 <= overall_pick <= 10:
        return '6-10'
    elif 11 <= overall_pick <= 15:
        return '11-15'
    else:
        return 'Other'

nba_data['draft_pick_category'] = nba_data['overall_pick'].apply(categorize_draft_pick)


performance_metrics_comparison = ['points_per_game', 'average_total_rebounds', 'average_assists']


draft_pick_performance = nba_data.groupby('draft_pick_category')[performance_metrics_comparison].mean()


relevant_categories = ['Top 5', '6-10', '11-15']
draft_pick_performance_filtered = draft_pick_performance.loc[relevant_categories]

draft_pick_performance_filtered

```

Iz ovih podataka nam je jasno da postoji velika razlika u kvaliteti između top 5 pickova i pickova 6-10, kao što postoji velika razlika između pickova 6-10 i 11-15.

**Korelacija između MVP kandidature i aktivnih godina igrača**

U ovom primjeru ćemo provjeriti koji je prosječni broj godina koje MVP kandidat provede u NBA ligi.

```python

mvp_candidates = df1['full_name'].unique()

years_active_info = df[df['player'].isin(mvp_candidates)][['player', 'years_active']]

average_years_active = years_active_info['years_active'].mean()

import matplotlib.pyplot as plt

plt.hist(years_active_info['years_active'], bins=range(0, int(years_active_info['years_active'].max()) + 1, 1), color='blue', alpha=0.7)
plt.title('Distribution of Years Active Among MVP Candidates')
plt.xlabel('Years Active in the NBA')
plt.ylabel('Number of MVP Candidates')
plt.grid(axis='y')
plt.show()

print(f"Average years active among MVP candidates: {average_years_active:.2f}")
```

Prosječan broj aktivnih godina za MVP kandidata je 14.93. Prema distribuciji podataka u grafu vidljiva je izrazita lijeva iskrivljenost grafa što nam jasno sugerira da postoji jaka povezanost između broja aktivnih godina igrača i MVP kandidature. Riječ je o poprilično intuitivnom zaključku s obzirom da sjajni igrači i u pozamašnim igračkim godinama sa svojim talentom i iskustvom mogu i dalje kontribuirati na najvišoj razini.


**MVP KANDIDATURA I DOB KANDIDATA**

U idućoj analizi ćemo istražiti nakon koliko godina karijere igrač ima najvišu šansu postati MVP kandidat. S obzirom da se u košrakaškim krugovima razdoblje između 27. i 32. godine smatra vrhuncem igrača s obizrom na njegovo postignuto iskustvo i fizičku spremu, predviđam da će prosjek biti između 6 i 8 godina aktivne karijere.

```python
player_draft_year = df[['player', 'year']].drop_duplicates(subset='player').set_index('player')

df1['draft_year'] = df1['full_name'].map(player_draft_year['year'])
df1['years_since_draft'] = df1['year'] - df1['draft_year']  

mvp_years_since_draft = df1.dropna(subset=['years_since_draft'])[['full_name', 'years_since_draft']]

average_years_since_draft = mvp_years_since_draft['years_since_draft'].mean()
median_years_since_draft = mvp_years_since_draft['years_since_draft'].median()
print(f"Average years since being drafted for MVP candidates: {average_years_since_draft:.2f}")
print(f"Median years since being drafted for MVP candidates: {median_years_since_draft:.2f}")
```

Iz reultata iznad primjećujemo da će prosječni MVP kandidat postati kandidat nakon 6 godina aktivne NBA karijere. Riječ je o očekivanom rezultatu s obzirom da u tom razdoblju igrač ima popriličnu količinu iskustva u NBA ligi te se nalazi na vrhuncu svoje fizičke spreme.

**Broj MVP kandidata po timu**

U ovoj analizi ćemo pogledati koliko su MVP kandidata imali NBA timovi u razdoblju između 1989 i 2021. 

```python
mvp_candidates = df1['full_name'].unique()

team_info = df[df['player'].isin(mvp_candidates)][['player', 'team']]

team_frequency = team_info['team'].value_counts()

print(team_frequency)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(team_frequency.index, team_frequency.values, color='purple')
plt.title('Frequency of Drafted Teams Among MVP Candidates')
plt.xlabel('Team')
plt.ylabel('Number of MVP Candidates')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Iz ovog dataseta primjećujemo da su Golden State Warriorsi imali najviše draftiranih MVP kandidata. Općenito primjećujemo da su timovi sa više draftiranih MVP kandidata imali veći uspjeh kroz povijest NBA lige. No, ova analiza može navesti na pojednica na krivi zaključak te ćemo kao primjer uzeti momčad Minesotta Timberwolvsa. Riječ je o jednoj od najneuspješnijih NBA franšiza koja je unatoč svemu imala 4 MVP kandidata. Ovo možemo pribrojiti različitim faktorima poput lošeg vodstva kluba, nekvalitetne strukture rostera i teškim pribavljanjem lukrativnih free agenta. Također, kod All star igrača u nekompetitivnim timovima često dolazi do inflacije statistike dovodi do boljeg učinka podjedinca i veće šanse za MVP kandidaturu, no ne garantira uspjeh tima u regularnoj sezoni i playoffu.

**Usporedba MVP kandidata i ostalih igrača**

```python
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, Float, String
import pandas as pd


df = pd.read_csv('nbaplayersdraft.csv')


engine = create_engine('sqlite:///nba_players_draft.db') 


metadata = MetaData(engine)


nba_draft_table = Table('nba_draft', metadata,
    Column('id', Integer, primary_key=True),
    Column('year', Integer),
    Column('rank', Integer),
    Column('overall_pick', Integer),
    Column('team', String),
    Column('player', String),
    Column('college', String),
    Column('years_active', Integer),
    Column('games', Integer),
    Column('minutes_played', Float),
    Column('field_goal_percentage', Float),
    Column('3_point_percentage', Float),
    Column('free_throw_percentage', Float),
    Column('average_minutes_played', Float),
    Column('points_per_game', Float),
    Column('average_total_rebounds', Float),
    Column('average_assists', Float),
    Column('win_shares', Float),
    Column('win_shares_per_48_minutes', Float),
    Column('box_plus_minus', Float),
    Column('value_over_replacement', Float)
)


metadata.create_all()


df.to_sql('nba_draft', con=engine, if_exists='replace', index=False)  
```

## **POSTAVLJANJE API SERVERA FLASK**

```python
pip install flask
```

```python
from flask import Flask, jsonify
from sqlalchemy import create_engine, MetaData, Table

app = Flask(__name__)


engine = create_engine('sqlite:///nba_players_draft.db')
metadata = MetaData(bind=engine)
nba_draft = Table('nba_draft', metadata, autoload=True)

@app.route('/players', methods=['GET'])
def get_players():
    conn = engine.connect()
    query = nba_draft.select()
    result = conn.execute(query)
    players = [dict(row) for row in result]
    conn.close()
    return jsonify(players)

if __name__ == '__main__':
    app.run(host='0.0.0.0')  

```
